# CONVERTING TIME INTO DESIRED FORMAT

```python
final=[]
for x in range (len(df['time'])):
  y=df1[x]
  a=time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(y))
  b=a[2:19]
  final.append(b)
```
